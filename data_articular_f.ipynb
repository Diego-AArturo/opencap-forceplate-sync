```python
import pandas as pd
import re

# ──────────────────── 1. Cargar archivos ────────────────────
bd = pd.read_excel('tablas_af/Tablas_BD_post.xlsx', sheet_name='EstPostIz')   # %BW
nw = pd.read_excel('tablas_af/Tablas_NW_post.xlsx', sheet_name='EstPostIz')   # N

# ──────────────────── 2. Función de normalización de nombres ────────────────────
def normalizar_columnas(df):
    ren = {}
    for col in df.columns:
        s = col.upper().replace('-', '').replace(' ', '')
        # RODSAGITAL  (ej.: F1RODSAGITALD  ó  F1_RODSAGITALD  ó  F1-RODSAGITAL D)
        m = re.match(r'(F\d)_?RODSAGITAL([DI])', s)
        if m:
            fase   = m.group(1)         # F1, F2, F3...
            pierna = m.group(2)         # D o I
            ren[col] = f'{fase}-RODSAGITAL {pierna}'
        else:
            ren[col] = col   # sin cambios
    return df.rename(columns=ren)
# print('1 ',bd.columns[:10])
bd = normalizar_columnas(bd)
nw = normalizar_columnas(nw)
# print('2 ',bd.columns[:10])
# ──────────────────── 3. Añadir unidad a columnas de fuerza ────────────────────
force_regex = r'^F(?!\d*TIME)(?!\d*-?RODSAGITAL).*'   # look-ahead negativo
fuerza_bd = bd.filter(regex=force_regex).columns
fuerza_nw = nw.filter(regex=force_regex).columns

bd = bd.rename(columns={c: f'{c}(BD)' for c in fuerza_bd})   # Body-weight (%BW)
nw = nw.rename(columns={c: f'{c}(NW)' for c in fuerza_nw})   # Newton (N)
# print('3 ',bd.columns)
# ──────────────────── 4. Eliminar RODSAGITAL duplicados de la tabla NW ────────────────────
rod_cols = [f'F{i}-RODSAGITAL {p}' for i in (1, 2, 3) for p in ('D', 'I')]
nw = nw.drop(columns=[c for c in rod_cols if c in nw.columns])

# ──────────────────── 5. Unir DataFrames ────────────────────
# Si el peso coincide exactamente:
common_cols = ['PESO', 'F1TIME','F2TIME','F3TIME']  # Puedes incluir F2TIME o F3TIME si existen

df_final = (
    bd
    .merge(nw.drop(columns=[c for c in common_cols if c in nw.columns]), on=['Sujeto'], how='inner')
    .sort_values('Sujeto')
    .reset_index(drop=True)
)


# ► df_final ahora contiene:
#   • Una sola columna para cada RODSAGITAL (F1, F2, F3; D / I).
#   • Todas las columnas de fuerza dos veces: “… (BD)” y “… (NW)”.
#   • Cada fila corresponde al mismo sujeto y peso.

# ──────────────────── 6. Guardar o inspeccionar ────────────────────
# print('final ',df_final.columns)
# df_final
```

```python
df = df_final

df_mod = df.copy()

# Ejes y fases a procesa

df_stas= df_mod.describe().loc[['mean','std']].T
df_stas['cv'] = (df_stas['std'] / np.abs(df_stas['mean'])).round(2)
fases = ['F1', 'F2', 'F3']
df_stas = df_stas.T
df_stas.drop(columns=[f"{fase}TIME" for fase in fases], inplace=True)
df_stas.drop(columns=['PESO'], inplace=True)
df_stas = df_stas.T
df_stas['media(ds)'] = df_stas['mean'].round(3).astype(str) + ' (±' + df_stas['std'].round(2).astype(str) + ')'
```

```python
import re
import numpy as np
import pandas as pd

# Mapeo de la cadena del eje → nombre legible
eje_map = {
    'compresivas'    : 'Compresivas',
    'anteroposteior' : 'Anteroposteior',
    'medilaterales'  : 'Medio-lateral'
}

# Unidades
unidad_map = {'BD': 'BW', 'NW': 'N'}

# Expresión regular para desmenuzar el índice
#  Ej.  FcompresivasF_T1(BD)  o  FanteroposteiorP_1_T3(NW)
pat = re.compile(
    r'^F(?P<eje>compresivas|anteroposteior|medilaterales)'
    r'(?P<tipo>[FP])'               # F o P
    r'(?P<pierna>_1)?'                      # el “_1” (pierna I) si aparece
    r'_T(?P<trial>\d)'              # número de ensayo
    r'\((?P<unidad>BD|NW)\)$',
    flags=re.I
)
def build_formatted(df_stats: pd.DataFrame, tipo: str) -> pd.DataFrame:
    """
    Devuelve una tabla estilo formatted_df_3
    solo con fuerzas del tipo 'F' o 'P'.
    """
    filas = []
    fases = ['F1', 'F2', 'F3']
    # Rodillas por fase
    rod_d = {f: df_stats.loc[f'{f}-RODSAGITAL D', 'media(ds)'] for f in fases}
    rod_i = {f: df_stats.loc[f'{f}-RODSAGITAL I', 'media(ds)'] for f in fases}

    # Agrupación de fuerzas por clave (fase, eje, unidad)
    grouped = {}
    for idx in df_stats.index:
        m = pat.match(idx)
        if not m or m['tipo'].upper() != tipo.upper():
            continue

        fase = f"F{m['trial']}"
        eje  = eje_map[m['eje'].lower()]
        met  = unidad_map[m['unidad']]
        pierna = 'I' if m['pierna'] else 'D'

        clave = (fase, eje, met)
        if clave not in grouped:
            grouped[clave] = {}
        grouped[clave][pierna] = {
            'media(ds)': df_stats.loc[idx, 'media(ds)'],
            'cv': df_stats.loc[idx, 'cv']
        }

    # Construcción de filas formateadas
    for (fase, eje, met), datos in grouped.items():
        rd_val = datos.get('D', {}).get('media(ds)', '')
        ri_val = datos.get('I', {}).get('media(ds)', '')
        rd_cv  = datos.get('D', {}).get('cv', np.nan)
        ri_cv  = datos.get('I', {}).get('cv', np.nan)

        filas.append({
            'FASE': fase,
            'RDM': rod_d[fase],
            'RIM': rod_i[fase],
            'EJE': eje,
            'MÉTRICA': met,
            'RDM_GR': rd_val,
            'RIM_GR': ri_val
        })

        filas.append({
            'FASE': '', 'RDP': '', 'RIP': '', 'EJE': '',
            'MÉTRICA': 'CV',
            'RDM_GR': f"{rd_cv*100:.2f}%" if not np.isnan(rd_cv) else '',
            'RIM_GR': f"{ri_cv*100:.2f}%" if not np.isnan(ri_cv) else ''
        })

    cols = ['FASE', 'RDM', 'RIM', 'EJE', 'MÉTRICA', 'RDM_GR', 'RIM_GR']
    return pd.DataFrame(filas, columns=cols).reset_index(drop=True)

formatted_F = build_formatted(df_stas, 'F')   # fuerzas F
formatted_P = build_formatted(df_stas, 'P')   # fuerzas P

print('Tabla fuerzas F')
display(formatted_F)

print('Tabla fuerzas P')
display(formatted_P)

```

```python
import re

# ---------- utilidades ----------
def parse_val_sd(texto):
    """
    Extrae media y desviación estándar de un string tipo '0.51 (±0.10)'
    """
    m = re.match(r'\s*([+-]?\d+(?:\.\d+)?)\s*\(±\s*([+-]?\d+(?:\.\d+)?)\s*\)', str(texto))
    return (float(m.group(1)), float(m.group(2))) if m else (np.nan, np.nan)

def fmt(mean, sd, dec=2):
    """Devuelve 'media (±sd)' con n decimales."""
    return f"{mean:.{dec}f} (±{sd:.{dec}f})"

def fmt_cv(cv):               # cv en fracción → porcentaje
    return f"{cv*100:.2f}%"

# ---------- parámetros ----------
fases        = ['F1', 'F2', 'F3']
ejes_cod     = {'VGRFY': 'Vertical', 'VGRFX': 'Anteroposterior', 'VGRFZ': 'Medio-lateral'}
piernas      = {'D': 'RD', 'I': 'RI'}
unidades_map = {'(%BW)': 'BW', '(NEWTON)': 'N'}

# ---------- generación ----------
filas = []
for fase in fases:
    # rodillas en sagital
    rd_rod = df_stas.loc[f"{fase}-RODSAGITAL D", 'media(ds)']
    ri_rod = df_stas.loc[f"{fase}-RODSAGITAL I", 'media(ds)']

    # recorre cada eje
    for cod_eje, nombre_eje in ejes_cod.items():
        # -------------------- BW y N --------------------
        for sufijo, etiqueta in unidades_map.items():
            # valores por pierna
            rd_idx = f"{fase}-{cod_eje} D {sufijo}"
            ri_idx = f"{fase}-{cod_eje} I {sufijo}"
            

            rd_val = df_stas.loc[rd_idx, 'media(ds)'] if rd_idx in df_stas.index else None
            ri_val = df_stas.loc[ri_idx, 'media(ds)'] if ri_idx in df_stas.index else None

            # calcula total y diferencia si faltan
            # if rd_val and ri_val:
            #     m_rd, s_rd = parse_val_sd(rd_val)
            #     m_ri, s_ri = parse_val_sd(ri_val)

            #     if not tot_val:
            #         tot_val = fmt(m_rd + m_ri, (s_rd**2 + s_ri**2) ** 0.5)
            #     if not diff_val:
            #         diff_val = fmt(abs(m_rd - m_ri), (s_rd**2 + s_ri**2) ** 0.5)

            # ---------- fila BW / N ----------
            filas.append({
                'FASE': fase,
                'RD': rd_rod,
                'RI': ri_rod,
                'EJE': nombre_eje,
                'MÉTRICA': etiqueta,
                'RD_GR': rd_val,
                'RI_GR': ri_val,
            })

            # ---------- fila CV ----------
            rd_cv = df_stas.loc[rd_idx, 'cv'] if rd_idx in df_stas.index else np.nan
            ri_cv = df_stas.loc[ri_idx, 'cv'] if ri_idx in df_stas.index else np.nan
            filas.append({
                'FASE': '',
                'RD': '',
                'RI': '',
                'EJE': '',
                'MÉTRICA': 'CV',
                'RD_GR': fmt_cv(rd_cv) if not np.isnan(rd_cv) else '',
                'RI_GR': fmt_cv(ri_cv) if not np.isnan(ri_cv) else '',
            })

# ---------- DataFrame final ----------
cols = ['FASE', 'RD', 'RI', 'EJE', 'MÉTRICA',
        'RD_GR', 'RI_GR']
formatted_df_3 = pd.DataFrame(filas, columns=cols)

# opcional: elimina filas con métricas vacías si fuera necesario
formatted_df_3 = formatted_df_3.dropna(subset=['RD_GR', 'RI_GR'], how='all').reset_index(drop=True)
formatted_df_3 
```
