Solo en los deslizamientos se debe canbiar r_ por L_ y l_ por R_

```python
def replace_words(input_file: str, output_file: str, switch: bool = False):
  '''
  input_file: Ruta al archivo de entrada.
  output_file: Ruta al archivo de salida.
  switch: Si es True, reemplaza 'gr' por 'ground' y 'r_' por 'L_' y 'l_' por 'R_'.
  '''
  # Verifica si el archivo de entrada existe
  if not os.path.exists(input_file):
      raise FileNotFoundError(f"El archivo {input_file} no existe.")

  # Diccionario de reemplazos
  if switch == True:
    replacements = {
        'gr': 'ground',
        'r_': 'L_',
        'l_': 'R_'
    }
  else:
    replacements = {
        'gr': 'ground',
        'r_': 'R_',
        'l_': 'L_'
    }

  try:
      # Lee el archivo de entrada
      with open(input_file, encoding='latin-1') as f:
          table = f.read()

      # Reemplaza las palabras
      for old, new in replacements.items():
          table = table.replace(old, new)

      # Escribe el archivo modificado en el archivo de salida
      with open(output_file, 'w', encoding='latin-1') as file:
          file.write(table)

      # print(f"El archivo ha sido modificado y guardado como {output_file}")


  except Exception as e:
      print(f"Ocurrió un error: {e}")
```
